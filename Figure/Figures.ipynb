```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

# Simulated data

```python
# PHF, DPER, Mice, MissForest, Soft-Impute
method = ['PHF','DPER','MICE','MF','SI']
err_cov = np.array([[0.237, 0.237, 0.434, 0.347, 0.79],
       [0.263, 0.262, 0.752, 0.607, 1.180],
       [0.286, 0.285, 1.084, 0.892, 1.315],
       [0.314, 0.312, 1.358 , 1.124, 1.268]])

err_inv_cov = np.array([[0.252, 0.252, 0.4, 0.327, 0.502],
       [0.283, 0.282, 0.735, 0.496, 0.625],
       [0.315, 0.314, 1.358, 0.668, 0.659],
       [0.346, 0.344, 2.392, 0.791, 0.648]])
```

```python
bar_width = 0.35
bar_positions1 = np.arange(len(method))
bar_positions2 = bar_positions1 + bar_width
plt.figure(figsize=(16, 6))

plt.subplot(141)
ax = plt.gca()
ax.bar(bar_positions1, err_cov[0], color = "blue", width = 0.35, label = r'$||\mathbf{\Sigma} - \mathbf{\hat{\Sigma}} ||$')
ax.bar(bar_positions2, err_inv_cov[0], width = 0.35, color = "red",  label = r'$||\mathbf{\Sigma^{-1}} - \mathbf{\hat{\Sigma}^{-1}} ||$')
plt.ylabel('Error',fontsize = 13)
plt.title('20% missing rate',fontsize = 13)
plt.xticks(bar_positions1 + bar_width / 2, method)
ax.legend()

plt.subplot(142)
ax = plt.gca()
ax.bar(bar_positions1, err_cov[1], color = "blue", width = 0.35, label = r'$||\mathbf{\Sigma} - \mathbf{\hat{\Sigma}} ||$')
ax.bar(bar_positions2, err_inv_cov[1], width = 0.35, color = "red",  label = r'$||\mathbf{\Sigma^{-1}} - \mathbf{\hat{\Sigma}^{-1}} ||$')
plt.title('40% missing rate',fontsize = 13)
plt.xticks(bar_positions1 + bar_width / 2, method)

plt.subplot(143)
ax = plt.gca()
ax.bar(bar_positions1, err_cov[2], color = "blue", width = 0.35, label = r'$||\mathbf{\Sigma} - \mathbf{\hat{\Sigma}} ||$')
ax.bar(bar_positions2, err_inv_cov[2], width = 0.35, color = "red",  label = r'$||\mathbf{\Sigma^{-1}} - \mathbf{\hat{\Sigma}^{-1}} ||$')
plt.title('60% missing rate',fontsize = 13)
plt.xticks(bar_positions1 + bar_width / 2, method)

plt.subplot(144)
ax = plt.gca()
ax.bar(bar_positions1, err_cov[3], color = "blue", width = 0.35, label = r'$||\mathbf{\Sigma} - \mathbf{\hat{\Sigma}} ||$')
ax.bar(bar_positions2, err_inv_cov[3], width = 0.35, color = "red",  label = r'$||\mathbf{\Sigma^{-1}} - \mathbf{\hat{\Sigma}^{-1}} ||$')
plt.title('80% missing rate',fontsize = 13)
plt.xticks(bar_positions1 + bar_width / 2, method)

plt.show()
```

```python
missing_rate = ['20%','40%','60%','80%']
err_PHF = err_cov[:,0]
inv_err_PHF = err_inv_cov[:,0]
err_DPER = err_cov[:,1]
inv_err_DPER = err_inv_cov[:,1]
err_MICE = err_cov[:,2]
inv_err_MICE = err_inv_cov[:,2]
err_MF = err_cov[:,3]
inv_err_MF = err_inv_cov[:,3]
err_SI = err_cov[:,4]
inv_err_SI = err_inv_cov[:,4]
plt.figure(figsize=(16, 6))
plt.subplot(121)
plt.plot(missing_rate, err_PHF,marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#c1272d", label = 'PHF')
plt.plot(missing_rate, err_DPER,marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#0000a7",  label = 'DPER',alpha = .3)
plt.plot(missing_rate, err_MICE,marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#eecc16",  label = 'MICE')
plt.plot(missing_rate, err_MF,marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#008176",  label = 'MF')
plt.plot(missing_rate, err_SI,marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#b3b3b3",  label = 'SI')
plt.ylabel('Error',fontsize = 13)
plt.xlabel('Missing rate',fontsize = 13)
plt.title('Covariance matrix estimation error',fontsize = 13)
plt.legend()

plt.subplot(122)
plt.plot(missing_rate, inv_err_PHF,marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#c1272d", label = 'PHF')
plt.plot(missing_rate, inv_err_DPER,marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#0000a7",  label = 'DPER',alpha = .3)
plt.plot(missing_rate, inv_err_MICE,marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#eecc16",  label = 'MICE')
plt.plot(missing_rate, inv_err_MF,marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#008176",  label = 'MF')
plt.plot(missing_rate, inv_err_SI,marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#b3b3b3",  label = 'SI')
plt.ylabel('Error',fontsize = 13)
plt.xlabel('Missing rate',fontsize = 13)
plt.title('Precision matrix estimation error',fontsize = 13)
plt.legend()

plt.show()
```
